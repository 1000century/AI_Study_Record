```python
from bs4 import BeautifulSoup

PAGE_NUM = 20
# HTML 파싱
soup = BeautifulSoup(docs[PAGE_NUM].page_content, "html.parser")

# 좌표 리스트
coordinates = docs[PAGE_NUM].metadata["coordinates"]

# 각 태그를 id 순으로 돌면서 매핑
for idx, tag in enumerate(soup.find_all(id=True)):
    tag_id = idx
    if tag_id < len(coordinates):
        print(f"[ID: {tag_id}] Text: {tag.get_text(strip=True)}")
        print(f"   BBox: {coordinates[tag_id]}")
    else:
        print(f"[ID: {tag_id}] has no coordinate info")


from pdf2image import convert_from_path

# PDF 경로
pdf_path = r"C:\Users\Sese\PROJECT\TAWILWIND\medihelper\private_data\snuh\nonOCR\snuh_1.pdf"

# 첫 페이지만 이미지로 변환
pages = convert_from_path(pdf_path, dpi=150, first_page=PAGE_NUM+1, last_page=PAGE_NUM+1)
page_image = pages[0]

# 이미지 저장
page_image.save("page1.png")


from PIL import ImageDraw
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup

# 좌표 정보와 HTML
# 이미지 크기
img_w, img_h = page_image.size
draw = ImageDraw.Draw(page_image)

# 박스 그리기
for idx, tag in enumerate(soup.find_all(id=True)):
    tag_id = idx
    text = tag.get_text(strip=True)

    if tag_id < len(coordinates):
        box = coordinates[tag_id]
        if box:
            x_coords = [pt["x"] * img_w for pt in box]
            y_coords = [pt["y"] * img_h for pt in box]
            min_x, max_x = min(x_coords), max(x_coords)
            min_y, max_y = min(y_coords), max(y_coords)

            draw.rectangle([min_x, min_y, max_x, max_y], outline="red", width=2)
            draw.text((min_x, min_y - 10), text[:20], fill="blue")  # 텍스트 일부만

# 결과 보여주기
plt.figure(figsize=(12, 8))
plt.imshow(page_image)
plt.axis("off")
plt.show()

```
