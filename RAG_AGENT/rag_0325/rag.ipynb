```python
# 현재 디렉토리 출력

import os
from typing import List, Optional, Any
from langchain_core.documents import Document

print(os.getcwd())
```

```python
from rag import RetrievalChain

from dotenv import load_dotenv
load_dotenv()
```

```python
VECTOR_DIR = "vector_db"
source_folder = <source_folder>
```

```python
rag_chain = RetrievalChain(
    source_url = [os.path.join(source_folder, i) for i in os.listdir(source_folder)],    
    persist_directory = str(VECTOR_DIR),
    k = 5,
    embedding_model = "text-embedding-3-small",
    llm_model = "gpt-4o-mini"
)

def format_search_results(docs:List[Document]) -> str:
    if not docs:
        return "No results found."
    markdown_results =  "## Search Results\n\n"
    for i, doc in enumerate(docs):
        source = doc.metadata.get('source','Unknown source')
        source = source.split("\\")[-1]
        page = doc.metadata.get('page', None)
        page_info = f" (page {page})" if page is not None else ""
        
        markdown_results += f'### Result {i}{page_info}\n\n'
        markdown_results += f"{doc.page_content}\n\n"
        markdown_results += f"Source: {source}\n\n"
        markdown_results += f"---\n\n"
        
    return markdown_results

```

```python
rag_chain.initialize()
```

```python
results = rag_chain.search_semantic('심실빈맥의 치료')
print(format_search_results(results))
```

```python
results = rag_chain.search_semantic('당뇨병의 진단 기준')
print(format_search_results(results))
```

```python
results = rag_chain.search_keyword('당뇨병의 진단 기준')
print(format_search_results(results))
```
